Feature Engineering

```python
import pandas as pd
import numpy as np

# Load dataset used in EDA (cleaned version)
df = pd.read_csv("online_shoppers_intention.csv")

print("Dataset loaded successfully")
print("Shape:", df.shape)
print("Target distribution:\n", df['Revenue'].value_counts())
```

```python
# Total pages viewed
df['Total_Pages_Viewed'] = (
    df['Administrative'] +
    df['Informational'] +
    df['ProductRelated']
)

print(" Total_Pages_Viewed created")
print(df['Total_Pages_Viewed'].head())
```

```python
# Total time spent
df['Total_Time_Spent'] = (
    df['Administrative_Duration'] +
    df['Informational_Duration'] +
    df['ProductRelated_Duration']
)

print(" Total_Time_Spent created")
print(df['Total_Time_Spent'].head())
```

```python
df['Avg_Time_Per_Page'] = (
    df['Total_Time_Spent'] / (df['Total_Pages_Viewed'] + 1)
)

print(" Avg_Time_Per_Page created")
print(df[['Total_Time_Spent', 'Total_Pages_Viewed', 'Avg_Time_Per_Page']].head())
print("Feature created: Avg_Time_Per_Page (time quality metric)")
```

```python
df['Product_Page_Ratio'] = df['ProductRelated'] / (df['Total_Pages_Viewed'] + 1)

print("Feature created: Product_Page_Ratio")
print(df[['ProductRelated', 'Total_Pages_Viewed', 'Product_Page_Ratio']].head())    
```

```python
df['Product_Time_Ratio'] = df['ProductRelated_Duration'] / (df['Total_Time_Spent'] + 1)

print("Feature created: Product_Time_Ratio")
print(df[['ProductRelated_Duration', 'Total_Time_Spent', 'Product_Time_Ratio']].head()) 
```

```python
df['High_Bounce'] = (df['BounceRates'] > 0.2).astype(int)

print("Binary feature created: High_Bounce")
print(df[['BounceRates', 'High_Bounce']].head())
```

```python
df['High_Exit'] = (df['ExitRates'] > 0.2).astype(int)

print("Binary feature created: High_Exit")
print(df[['ExitRates', 'High_Exit']].head())
```

```python
df['Discount_Active'] = (df['SpecialDay'] > 0).astype(int)

print("Feature created: Discount_Active")
print(df[['SpecialDay', 'Discount_Active']].sample(28))
```

```python
df['Discount_Level'] = pd.cut(
    df['SpecialDay'],
    bins=[-0.01, 0, 0.5, 1],
    labels=['No_Discount', 'Medium_Discount', 'High_Discount']
)

print("Feature created: Discount_Level (categorical)")
print(df[['SpecialDay', 'Discount_Level']].sample(28))
```

```python
df['Is_Returning_Visitor'] = (df['VisitorType'] == 'Returning_Visitor').astype(int)

print("Binary feature created: Is_Returning_Visitor")
print(df[['VisitorType', 'Is_Returning_Visitor']].sample(28))
```

```python
df['Weekend_Flag'] = df['Weekend'].astype(int)

print("Weekend_Flag ready for modeling")
print(df[['Weekend', 'Weekend_Flag']].sample(6))
```

```python
df['Month'] = pd.Categorical(
    df['Month'],
    categories=[
        'Jan','Feb','Mar','Apr','May','June',
        'Jul','Aug','Sep','Oct','Nov','Dec'
    ],
    ordered=True
)

df['Month_Code'] = df['Month'].cat.codes

print("Feature created: Month_Code (seasonality)")
print(df[['Month', 'Month_Code']].sample(6)) 
```

```python
df['High_Page_Value'] = (df['PageValues'] > 0).astype(int)

print("Feature created: High_Page_Value")
print(df[['PageValues', 'High_Page_Value']].sample(6))
```

```python
df['Discount_Product_Interaction'] = (
    df['Discount_Active'] * df['Product_Page_Ratio']
)

print("Feature created: Discount_Product_Interaction")
print(df[['Discount_Active', 'Product_Page_Ratio', 'Discount_Product_Interaction']].sample(28))
```

```python
df['Returning_Engagement'] = (
    df['Is_Returning_Visitor'] * df['Total_Time_Spent']
)

print("Feature created: Returning_Engagement")
print(df[['Is_Returning_Visitor', 'Total_Time_Spent', 'Returning_Engagement']].sample(6))
```

```python
df['Engagement_Level'] = pd.qcut(
    df['Total_Time_Spent'],
    q=3,
    labels=['Low_Engagement', 'Medium_Engagement', 'High_Engagement']
)

print(" Engagement_Level created")
print(df['Engagement_Level'].value_counts())
print(df[['Total_Time_Spent', 'Engagement_Level']].sample(6))
```

```python
df['Hesitation_Score'] = (
    df['Total_Pages_Viewed'] / (df['Avg_Time_Per_Page'] + 1)
)

print(" Hesitation_Score created")
print(df['Hesitation_Score'].describe())
print(df[['Total_Pages_Viewed', 'Avg_Time_Per_Page', 'Hesitation_Score']].sample(6))
```

```python
df['Discount_Sensitive_User'] = (
    (df['Discount_Active'] == 1) &
    (df['Product_Page_Ratio'] > 0.6)
).astype(int)

print(" Discount_Sensitive_User created")
print(df['Discount_Sensitive_User'].value_counts())
print(df[['Discount_Active', 'Product_Page_Ratio', 'Discount_Sensitive_User']].sample(28))
```

```python
df['Traffic_Quality_Score'] = (
    df['Total_Time_Spent'] *
    df['Product_Page_Ratio'] *
    (1 - df['BounceRates'])
)

print(" Traffic_Quality_Score created")
print(df['Traffic_Quality_Score'].describe())
print(df[['Total_Time_Spent', 'Product_Page_Ratio', 'BounceRates', 'Traffic_Quality_Score']].sample(6))
```

```python
drop_cols = [
    'Administrative', 'Informational', 'ProductRelated',
    'Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration',
    'VisitorType', 'Weekend', 'Month'
]

df_fe = df.drop(columns=drop_cols)

print("Dropped redundant raw columns")
print("Final feature set shape:", df_fe.shape)
```

```python
print("Final feature columns:")
print(df_fe.columns)

print("\nSample engineered data:")
print(df_fe.head())
```

```python
corr = df_fe.corr(numeric_only=True)['Revenue'].sort_values(ascending=False)
print(corr)
```

```python
drop_cols_step1 = [
    'PageValues',
    'High_Page_Value',
    'Browser',
    'TrafficType',
    'Region',
    'OperatingSystems'
]

df_fe = df_fe.drop(columns=drop_cols_step1)

print("STEP 1 DONE: Leakage & noise dropped")
print("Remaining columns:")
print(df_fe.columns)
print("Shape:", df_fe.shape)
```

```python
from sklearn.feature_selection import VarianceThreshold

# Select numeric features only (exclude target)
X_numeric = df_fe.drop(columns=['Revenue']).select_dtypes(include=['int64', 'float64'])

print("Numeric columns checked for variance:")
print(X_numeric.columnzs)
```

```python
selector = VarianceThreshold(threshold=0.01)
selector.fit(X_numeric)

low_variance_features = X_numeric.columns[
    ~selector.get_support()
]

print("Low variance features to drop:", list(low_variance_features))
```

```python
df_fe = df_fe.drop(columns=['BounceRates', 'ExitRates'])

print("STEP 2 DONE: Dropped low-variance features")
print("Remaining columns:")
print(df_fe.columns)
print("Shape:", df_fe.shape)
```

```python
skewness = df_fe.drop(columns=['Revenue']).skew(numeric_only=True).sort_values(ascending=False)

print("Skewness of numeric features:")
print(skewness)
```

```python
corr_matrix = df_fe.drop(columns=['Revenue']).corr(numeric_only=True)

# Find highly correlated feature pairs
high_corr_pairs = (
    corr_matrix.abs()
    .where(corr_matrix.abs() > 0.85)
    .stack()
    .reset_index()
)

high_corr_pairs.columns = ['Feature_1', 'Feature_2', 'Correlation']

# Remove self-correlation
high_corr_pairs = high_corr_pairs[high_corr_pairs['Feature_1'] != high_corr_pairs['Feature_2']]

print("Highly correlated feature pairs (>0.85):")
print(high_corr_pairs)
```

```python
drop_cols_step4 = [
    'SpecialDay',
    'Discount_Active',
    'Total_Pages_Viewed',
    'Total_Time_Spent'
]

df_fe = df_fe.drop(columns=drop_cols_step4)

print("STEP 4 DONE: Redundant features dropped")
print("Remaining columns:")
print(df_fe.columns)
print("Final shape:", df_fe.shape)
```

```python
print("Final columns:")
print(df_fe.columns)

print("\nShape:", df_fe.shape)
print("\nTarget distribution:")
print(df_fe['Revenue'].value_counts())
```

MODEL BUILDING

```python
from sklearn.model_selection import train_test_split

# Separate features and target
X = df_fe.drop(columns=['Revenue'])
y = df_fe['Revenue']

# Train-test split (stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    stratify=y,
    random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)

print("\nTarget distribution:")
print("Train:\n", y_train.value_counts(normalize=True))
print("Test:\n", y_test.value_counts(normalize=True))
```

```python
print(" TRAINTEST SPLIT SUMMARY (PRESENTATION FRIENDLY)")
print("="*55)

print(f"Total sessions in dataset: {len(X_train) + len(X_test)}")
print(f"Training set size: {len(X_train)} sessions ({len(X_train)/(len(X_train)+len(X_test))*100:.1f}%)")
print(f"Testing set size: {len(X_test)} sessions ({len(X_test)/(len(X_train)+len(X_test))*100:.1f}%)")

print("\n Purchase Distribution (Target Variable)")
print("-"*55)

train_purchase_rate = y_train.mean() * 100
test_purchase_rate = y_test.mean() * 100

print(f"Purchase rate in TRAIN set: {train_purchase_rate:.2f}%")
print(f"Purchase rate in TEST set:  {test_purchase_rate:.2f}%")

print("\n Interpretation:")
print(" The dataset is imbalanced (most users do not purchase).")
print(" Train and test sets have nearly identical purchase rates.")
print(" This confirms a correct stratified split.")
print(" The data is now ready for model training.")
```

```python
# Check data types of training features
X_train.dtypes
```

```python
from sklearn.preprocessing import OneHotEncoder

# Columns to encode
categorical_cols = ['Discount_Level', 'Engagement_Level']

# Initialize encoder
encoder = OneHotEncoder(drop='first', sparse_output=False)

# Fit on training data
encoded_train = encoder.fit_transform(X_train[categorical_cols])

print("Categorical features encoded (train)")
```

```python
# Get feature names
encoded_feature_names = encoder.get_feature_names_out(categorical_cols)

# Create DataFrame
encoded_train_df = pd.DataFrame(
    encoded_train,
    columns=encoded_feature_names,
    index=X_train.index
)

encoded_train_df.head()
```

```python
# Drop original categorical columns
X_train_encoded = X_train.drop(columns=categorical_cols)

# Concatenate encoded columns
X_train_encoded = pd.concat([X_train_encoded, encoded_train_df], axis=1)

print("Final X_train shape after encoding:", X_train_encoded.shape)
```

```python
# Transform test data using SAME encoder
encoded_test = encoder.transform(X_test[categorical_cols])

encoded_test_df = pd.DataFrame(
    encoded_test,
    columns=encoded_feature_names,
    index=X_test.index
)

X_test_encoded = X_test.drop(columns=categorical_cols)
X_test_encoded = pd.concat([X_test_encoded, encoded_test_df], axis=1)

print("Final X_test shape after encoding:", X_test_encoded.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize Logistic Regression
baseline_model = LogisticRegression(
    max_iter=1000,
    class_weight='balanced',
    random_state=42
)

# Train model on encoded data
baseline_model.fit(X_train_encoded, y_train)

print("Baseline Logistic Regression trained successfully")
```

```python
from sklearn.metrics import precision_score, recall_score, classification_report, confusion_matrix

# Make predictions on test set
y_pred = baseline_model.predict(X_test_encoded)

# Calculate metrics
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print(" BASELINE MODEL PERFORMANCE")
print("="*40)
print(f"Precision: {precision:.3f}")
print(f"Recall:    {recall:.3f}")

print("\nDetailed classification report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score

# Get predicted probabilities for class = 1 (purchase)
y_probs = baseline_model.predict_proba(X_test_encoded)[:, 1]

# Try a stricter threshold
threshold = 0.7
y_pred_thresh = (y_probs >= threshold).astype(int)

precision_t = precision_score(y_test, y_pred_thresh)
recall_t = recall_score(y_test, y_pred_thresh)

print(" THRESHOLD TUNING RESULTS")
print("="*40)
print(f"Threshold used: {threshold}")
print(f"Precision: {precision_t:.3f}")
print(f"Recall:    {recall_t:.3f}")
```

```python
for t in [0.5, 0.6, 0.7]:
    y_pred_t = (y_probs >= t).astype(int)
    p = precision_score(y_test, y_pred_t)
    r = recall_score(y_test, y_pred_t)
    print(f"Threshold {t}: Precision={p:.3f}, Recall={r:.3f}")
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize Random Forest (baseline, no tuning)
rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1,
    class_weight='balanced'
)

# Train the model
rf_model.fit(X_train_encoded, y_train)

print("Random Forest baseline model trained successfully")
```

```python
from sklearn.metrics import precision_score, recall_score, classification_report, confusion_matrix

# Predict on test set
y_pred_rf = rf_model.predict(X_test_encoded)

# Metrics
precision_rf = precision_score(y_test, y_pred_rf)
recall_rf = recall_score(y_test, y_pred_rf)

print(" RANDOM FOREST BASELINE PERFORMANCE")
print("="*45)
print(f"Precision: {precision_rf:.3f}")
print(f"Recall:    {recall_rf:.3f}")

print("\nDetailed classification report:")
print(classification_report(y_test, y_pred_rf))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
```

```python
from sklearn.metrics import precision_score, recall_score

# Get predicted probabilities for purchase class
rf_probs = rf_model.predict_proba(X_test_encoded)[:, 1]

print(" RANDOM FOREST  THRESHOLD TUNING")
print("="*45)

for t in [0.3, 0.4, 0.5]:
    y_pred_t = (rf_probs >= t).astype(int)
    p = precision_score(y_test, y_pred_t)
    r = recall_score(y_test, y_pred_t)
    print(f"Threshold {t}: Precision={p:.3f}, Recall={r:.3f}")
```
